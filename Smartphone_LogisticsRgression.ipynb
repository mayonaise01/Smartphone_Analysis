```python
# แก้ไขข้อมูลใน column 'หากท่านซื้อสมาร์ทโฟนเครื่องใหม่ จะใช้เลือกซื้อสมาร์ทโฟนแบรนด์ใด' โดยที่ถ้าข้อมูลใน column คือ 'Iphone' หรือ 'ยังไม่รู้' ให้คงค่านั้นไว้ นอกเหนือจาก 2 ค่านี้ให้ใส่ข้อมูลว่า 'not Iphone' แทน

df['หากท่านซื้อสมาร์ทโฟนเครื่องใหม่ จะใช้เลือกซื้อสมาร์ทโฟนแบรนด์ใด'] = df['หากท่านซื้อสมาร์ทโฟนเครื่องใหม่ จะใช้เลือกซื้อสมาร์ทโฟนแบรนด์ใด'].apply(lambda x: x if x in ['Iphone', 'ยังไม่รู้'] else 'not Iphone')
```

```python
from sklearn.preprocessing import LabelEncoder

# Select the target column
target_column = 'ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด'

# Create a LabelEncoder instance
le = LabelEncoder()

# Convert the target column to numerical values
df[target_column] = le.fit_transform(df[target_column])

# Show the mapping of classes to numerical values
class_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Class mapping:", class_mapping)

# Display the first few rows to see the changes
df.head(15)
```

```python
# Filter the DataFrame to show rows where 'ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด' equals 2
df[df['ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด'] == 2]
```

```python
# Filter the rows where 'ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด' is 2 (IOS และ Android)
df_both = df[df['ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด'] == 2]

# Create two new DataFrames: one with Android and one with iOS, duplicating the rows
df_android = df_both.copy()
df_android['ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด'] = 0  # Set the target to Android

df_ios = df_both.copy()
df_ios['ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด'] = 1  # Set the target to iOS

# Combine these new rows back with the original DataFrame (excluding the original rows with value 2)
df = pd.concat([df[df['ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด'] != 2], df_android, df_ios])

# Display the cleaned DataFrame with split rows
df.tail(10)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select the features
feature_columns = ['เพศของผู้ตอบแบบสอบถาม', 'ช่วงอายุของผู้ตอบแบบสอบถาม', 'ช่วงเงินเดือนของผู้ตอบแบบสอบถาม','ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [เล่นเกม]','ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [เล่นโซเชี่ยลมีเดีย]','ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [ดูหนัง ฟังเพลง]','ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [ถ่ายรูป ถ่ายวิดีโอ]','หากท่านซื้อสมาร์ทโฟนเครื่องใหม่ จะใช้เลือกซื้อสมาร์ทโฟนแบรนด์ใด']

# One-Hot Encoding of the categorical features
one_hot_encoder = OneHotEncoder(sparse=False)
X_encoded = one_hot_encoder.fit_transform(df[feature_columns])

# Create a DataFrame with the One-Hot Encoded features
X_encoded_df = pd.DataFrame(X_encoded, columns=one_hot_encoder.get_feature_names_out(feature_columns))

# Show the transformed features
X_encoded_df.head()
```

```python
df.head(15)
```

```python
X = X_encoded_df
Y = df['ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_X, train_Y)
```

```python
y_pred = model.predict(test_X)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(f"Accuracy: {accuracy_score(test_Y, y_pred)}")
print(classification_report(test_Y, y_pred))
```

```python
import shap
```

```python
explainer = shap.Explainer(model, train_X, feature_names=train_X.columns.tolist())
shap_values = explainer(test_X)
```
